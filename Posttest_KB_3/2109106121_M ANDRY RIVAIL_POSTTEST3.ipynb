```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Snapchat_app_store_reviews.csv')
df.head()
```

<h6>1.GRAFIK BATANG<h6>

```python
data_bar = df['rating'].value_counts()
```

```python
plt.bar(x=data_bar.index, height=data_bar.values)
plt.title('Total Rating')
```

<h6>2.HISTOGRAM<h6>

```python
plt.figure(figsize=(10, 6))
plt.hist(df['rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribusi Rating Snapchat')
plt.xlabel('Rating')
plt.ylabel('Jumlah Rating')
plt.show()
```

<h6>3.LINEPLOT<h6>

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df['date'])

plt.xticks(rotation=60)
plt.show()
```

<h6>4.HEATMAP<h6>

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGn')
plt.title('Heatmap Korelasai Antara Variabel Numerik')
plt.xlabel('Variabel')
plt.ylabel('Variabel')
plt.show()
```

<h6>5.PIE CHART<h6>

```python
data_pie = df['rating'].value_counts().rename('rating').reset_index(name='id_count') 
data_pie
```

```python
plt.pie(data_pie.id_count, labels=data_pie["index"], startangle=90,
autopct='%.1f%%', shadow= True,)
plt.title('rating')
plt.show()
```
